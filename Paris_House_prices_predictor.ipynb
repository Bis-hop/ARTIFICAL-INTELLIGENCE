```python
#importing all the required liberaries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
```

```python
#Path to the CSV file
test='/content/drive/MyDrive/ParisHousing.csv'
data= pd.read_csv(test)
#Displaying the shape of the data
print(data.shape)
#Displaying the data
print(data)
```

**Train Test Split**

```python
#Splits the dataset into training and testing sets.
X=data.drop('price',axis=1)#Extracts the features (independent variables) by removing the 'price' column.
#Represents the target variable (dependent variable), which is the 'price' column.
Y=data.price
#Test_0.3 allocates 30% of the data for testing.
X_train, X_test, Y_train,Y_test= train_test_split(X,Y,test_size=0.3, random_state=42)#Random_state gives initial state used in generating random numbers
```
